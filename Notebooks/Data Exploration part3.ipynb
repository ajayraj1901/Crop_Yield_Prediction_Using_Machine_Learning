Visualizing the data after data cleansing using Matplotlib and Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import PIL as pil
import random
random.seed(0)
np.random.seed(0)
%matplotlib inline
```

```python
ts_data=pd.read_csv("datasets/final_preprocessed_data.csv", encoding='utf-8')
ts_data.head()
```

```python
ts_data['Production_per_100000'] = ts_data['Production']/100000
ts_data['Area_per_100000'] = ts_data['Area']/100000
```

```python
unique_district = ts_data['Dist Name'].unique()
unique_season = ts_data['Season'].unique()
unique_crop = ts_data['Crop'].unique()
years = ts_data['Year'].unique()

print(len(unique_district))
print(unique_district)
```

```python
print(len(unique_crop))
print(unique_crop)
```

```python
print(len(unique_season))
print(unique_season)
```

```python
fig = plt.figure(figsize=(25,16))  
for i,year in enumerate(years):
    plt.subplot(6,3,i+1)
    plt.bar('Season','Production_per_100000',data=ts_data[ts_data.Year==year])
    plt.ylim(0,100)
    plt.text(0.5,50,str(year))
 
fig.suptitle("Production in every year in different Session")
fig.text(0.5,0,"Crop Season",fontsize=20)
fig.text(0,0.5,"Production_per_100000",rotation=90,fontsize=20) 
```

```python
group_data1 = ts_data.groupby(["Dist Name","Crop"])
production_data1 = group_data1['Production_per_100000'].sum()
area_data1 =  group_data1['Area_per_100000'].sum()
```
